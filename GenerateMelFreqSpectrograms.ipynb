```python
import pandas as pd
import numpy as np
import librosa as lb
import matplotlib.pyplot as plt
import os
import IPython.display as ipd
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
TRAIN_CSV = "Train.csv"
OUTPUT = "MelSpec"
```

```python
# Load train csv file
train_csv = pd.read_csv(TRAIN_CSV)
train_csv.head()
```

```python
# Create a folder named after each bird specie under the MelSpec folder. We will save the Mel Frequency Spectrograms from the audio files under their corresponding folders
for bird in train_csv['common_name'].unique():
        if not (os.path.exists(OUTPUT + rf"/{bird}")):
                os.makedirs(OUTPUT + rf"/{bird}")
```

```python
# Function that generates mel freqency spectrograms
def gen_spectrogram(input_path, output_path=None, save=True):
    
    x , sr = lb.load(input_path)
    X = lb.stft(x, n_fft=1024, hop_length=32, win_length=256)
    melspec = lb.feature.melspectrogram(y=None, S=X, sr=22050, n_fft=513, n_mels=128, htk=True, fmin=2000, fmax=11025)
    mel_DB = lb.power_to_db(melspec, ref=np.mean)
    
    fig = plt.figure(frameon=False)
    fig.set_size_inches(8, 8)
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
    ax.imshow(mel_DB, aspect='auto')
    if save == True:
        fig.savefig(output_path, dpi=512//8)
        ipd.clear_output(wait=True)
        plt.close()
    
```

```python
# Display a random mel feq spectrogram sample
idx = random.randint(0, len(train_csv))
sample_path = rf"Train/{train_csv['ID'][idx]}.mp3"
gen_spectrogram(sample_path, save=False)
```

```python
# Generate mel freq spec for all dataset and save them under their corresponding folder
counter = 1
print("We currently generating the melspec for: ", train_csv['common_name'][0], " - index: ", counter, " out of 40")
for i, row in train_csv.iterrows():
    if i < len(train_csv) - 1:
        if train_csv['common_name'][i] != train_csv['common_name'][i+1]:
            counter +=1
            print("We currently generating the melspec for: ", train_csv['common_name'][i+1], " - index: ", counter, " out of 40")
    gen_spectrogram(rf"Train/{row['ID']}.mp3", OUTPUT + rf"/{row['common_name']}/{row['ID']}.png")
```
