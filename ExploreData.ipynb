```python
import librosa as lb
import pandas as pd
import random as rd
from glob import glob
import IPython.display as ipd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
TRAIN_CSV = "Train.csv"
TEST_CSV = "Test.csv"
TRAIN = "Train/*.mp3"
TEST = "Test/*.mp3"
```

# Trainset

```python
train_csv = pd.read_csv(TRAIN_CSV)
train_csv.head()
```

=> This data frame contains the labels (name of the bird) of each audio file in the train folder.

```python
# Check nan values
print(f"Checking NaN values per feature (column): \n\n{train_csv.isnull().sum()}")
print(f"\nTotal NaN values: {sum(train_csv.isnull().sum())}")

print("\n*********************************\n")
# Number of audio samples
print(f"Number of audio samples in the train set: \n\n{len(train_csv)}")
print("\n*********************************\n")
# Number of classes (bird names)
print(f"Bird Species: \n\n{train_csv['common_name'].unique()}\n")
print(f"Number of Bird Species: \n\n{len(train_csv['common_name'].unique())}")
```

```python
# Read an audio sample from the train set
train = glob(TRAIN)
ipd.Audio(rd.choice(train))
```

=> As we execute the previous cell multiple times, we get a different random sample. \
=> The samples are different in terms of duration. 

```python
# Print the largest, smallest, and mean file duration
file_len = []
for i, row in train_csv.iterrows():
    # Load audio file
    pth = f"Train/{row['ID']}.mp3"
    audio_file, sr = lb.load(pth)
    file_len.append(len(audio_file))
print(f"Minimum file duration: {min(file_len)}")
print(f"Maximum file duration: {max(file_len)}")
print(f"Mean duration: {sum(file_len)/len(file_len)}")
```

### Visualize a sample per Specie

```python
# Form a dataframe that contains one sample per bird
train_viz = pd.DataFrame(columns=train_csv.columns)
for bird in train_csv['common_name'].unique():
    new_row = train_csv[train_csv['common_name'] == bird].sample(n = 1)
    train_viz = pd.concat([train_viz, new_row], ignore_index=True)
train_viz
```

```python
# Load audio samples in the previous dataframe and visualize them
fig, axs = plt.subplots(nrows=10, ncols=4, figsize=(40, 50))
axs = axs.ravel()
for file_id, bird, ax in zip(train_viz['ID'], train_viz['common_name'], range(40)):
    audio_file, sr = lb.load(rf"Train/{file_id}.mp3")
    axs[ax].plot(audio_file)
    axs[ax].set_title(bird)
fig.supxlabel('Time')
fig.supylabel('Amplitude')
plt.show()
```
